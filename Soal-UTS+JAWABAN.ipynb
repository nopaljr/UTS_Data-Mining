### 4
Create a function that return companies with list of employees that are working in the company.

```json

    "company": "Company 3",
```

````python
# lengkapi function berikut
def get_employees_by_company():
    company_detail_list = {
        'Company 1': {
            'name': 'Company 1',
            'domain': 'Retail',
            'clients': [
                {
                    'name': 'acme.inc',
                    'country': 'united states'
                },
                {
                    'name': 'Wayne.co',
                    'country': 'united states'
                }
            ]
        },
        'Company 2': {
            'name': 'Company 2',
            'domain': 'Construction',
            'clients': [
                {
                    'name': 'Tesla',
                    'country': 'united states'
                },
                {
                    'name': 'Japan Airlines',
                    'country': 'japan'
                },
                {
                    'name': 'Indofood',
                    'country': 'indonesia'
                }
            ]
        },
        'Company 3': {
            'name': 'Company 3',
            'domain': 'Healthcare',
            'clients': [
                {
                    'name': 'Petronas',
                    'country': 'malaysia'
                },
                {
                    'name': 'VW Group',
                    'country': 'germany'
                },
                {
                    'name': 'IBM',
                    'country': 'united states'
                },
                {
                    'name': 'Mitsubishi',
                    'country': 'japan'
                }
            ]
        }
    }

    employee_detail_list = {
        'John Doe': {
            'name': 'EMP-0001',
            'first_name': 'John',
            'last_name': 'Doe',
            'full_name': 'John Doe',
            'company': 'Company 1'
        },
        'Tom Smith': {
            'name': 'EMP-0002',
            'first_name': 'Tom',
            'last_name': 'Smith',
            'full_name': 'Tom Smith',
            'company': 'Company 2'
        },
        'Andrew Sebastian': {
            'name': 'EMP-0003',
            'first_name': 'Andrew',
            'last_name': 'Sebastian',
            'full_name': 'Andrew Sebastian',
            'company': 'Company 2'
        },
    }

    company_employees = {company: [] for company in company_detail_list.keys()}

    for employee in employee_detail_list.values():
        company_name = employee['company']
        company_employees[company_name].append(employee['full_name'])

    result_employee = []
    for company, employees in company_employees.items():
        result_employee.append({
            "company": company,
            "employees": employees
        })

    return result_employee

company_employee_list = get_employees_by_company()

for entry in company_employee_list:
    print(entry)
````

### Soal pre-processing data

````python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
import numpy as np
````

````python
data_startup = pd.read_csv('50_Startups.csv')
````

````python
data_startup.sample(7)
missing_data = data_startup.isnull().sum()
print("Data kosong:\n", missing_data)

# Mengisi data kosong dengan nilai mean
imputer = SimpleImputer(strategy='mean')
data_startup[['R&D Spend', 'Administration', 'Marketing Spend', 'Profit']] = imputer.fit_transform(data_startup[['R&D Spend', 'Administration', 'Marketing Spend', 'Profit']])

# One-hot encoding untuk kolom State
onehot_encoder = OneHotEncoder()
state_encoded = onehot_encoder.fit_transform(data_startup[['State']])
state_encoded_df = pd.DataFrame(state_encoded.toarray(), columns=onehot_encoder.get_feature_names_out(['State']))
data_startup = pd.concat([data_startup, state_encoded_df], axis=1)
data_startup.drop('State', axis=1, inplace=True)

# Membuat kolom baru Tax
data_startup['Tax'] = (data_startup['Profit'] + data_startup['Marketing Spend'] + data_startup['Administration']) * 0.05

# StandardScaler untuk semua fitur kecuali Profit
scaler = StandardScaler()
features_to_scale = [col for col in data_startup.columns if col != 'Profit']
data_startup[features_to_scale] = scaler.fit_transform(data_startup[features_to_scale])

# Menampilkan data setelah preprocessing
print("Data setelah preprocessing:\n", data_startup.head())
````

### Soal
Dari data di atas lakukan pre-processing sbb:
1. Lakukan analisa mengenai field mana saja yang memiliki data kosong(Nan) lalu Isi data yang kosong tersebut menggunakan nilai mean
2. lakukan oneHotEncoder ke field state
3. buat field baru yaitu Tax. dimana Tax = (profit + Marketing Spend + administration) * 5%
4. Jalankan standardScaler field-field tersebut
